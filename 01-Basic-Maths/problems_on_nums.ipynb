```python
# Check if a number is armstrong number or not
n = int(input())
temp ,sum = n,0
digits = len(str(n))
while n>0:
    digit = n % 10
    sum += digit ** digits
    n = n//10
if temp == sum:
    print("Amrstrong ")
else:
    print("Not a Amrstrong number")
```

```python
# Check if a number is perfect number
n = int(input())
sums = 0
for i in range(1,n):
    if n % i ==0:
        sums += i
if  sums == n:
    print("perfect")
else:
    print("Not")
```

```python
# Maximum and Minimum digit in a number
n = int(input())
max = 0
min = 9
while n > 0:
    digit = n% 10
    if digit > max:
        max = digit
    if digit < min:
        min = digit
    n//= 10
print(max)
print(min)
    
```

```python
# Check if a number is a strong number or not
n = int(input())
temp = n
tsum = 0
while n> 0:
    digit = n%10
    fact = 1
    for i in range(1,digit+1):
        fact *=i
    tsum +=fact
    n //=10
if tsum == temp:
    print("Strong Number")
else:
    print("Not A Strong Number")
```

```python
# LCM of two numbers
a, b = map(int,input().split())
oa , ob = a,b
while b:
    a,b =b,a%b
gcd = a
lcm = (oa * ob)//gcd
print("gcd",gcd)
print("lcm",lcm)
```
